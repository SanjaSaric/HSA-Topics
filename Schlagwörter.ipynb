Erstellt wird jeweils eine Liste (eine Datei = ein Listeneintrag). Der gleiche Index in zwei Listen indiziert Informationen für die gleiche Datei. Bei Stichwörtern einhält diese Liste auch Listen, die gemeinsam vorkommende Stichwörter pro Datei beinhalten. 

```python
senders = []
terms = []
ids = []

for path in path_to_folder.glob('*.xml'):
    with open(path, encoding="utf-8") as f:
        soup = BeautifulSoup(f, 'xml')
        term_group = []
        if soup.find('keywords', scheme="http://netknowl.uni-graz.at") != None:
            for term in (soup.find('keywords', scheme="http://netknowl.uni-graz.at")).find_all('term'): #, scheme="http://netknowl.uni-graz.at"
                if term.string is not None:
                    term_group.append(term.string)
            terms.append(term_group)
            #print(term_group)
            for sender in (soup.find('correspAction', type="sending")).find('persName'): #/TEI/teiHeader[1]/profileDesc[1]/correspDesc[1]/correspAction[1]/persName[1]
                senders.append(sender.string)
            for pid in soup.find('idno', type="PID"):
                ids.append(pid.split('.')[-1])
            
```

```python
# Stichwörter im Korpus auslesen und Zählen
def get_keywords(keywords):
    d = dict(zip(keywords, [0] * len(keywords)))
    for k in keywords:
        d[k] += 1
    return d

d = get_keywords(keywords)
for k in sorted(d, key=d.get, reverse=True):
    print(k, d[k])
```

```python
def get_tokens(root, keywords):
    tokens = []
    i = 0
    for k in keywords:
        if root.lower() in str(k).lower():
            i += 1
            tokens.append(k)
    return set(tokens), i
```

```python
# Die Metadaten aus den einzelnen Listen können auch in ein Dictionary gesammelt werden, um sie leichter ansprechen zu können

de_dict = []
counter = 0
for pid in ids_de:
    index = ids_de.index(pid)
    fileinfo = {'id':pid, 'sender':senders_de[index], 'terms':terms_de[index]}
    de_dict.append(fileinfo)
    counter +=1
            
```

```python
print(counter)
print(de_dict[:5])
```

```python
def get_sender_terms(sender):
    terms = []
    for item in de_dict:
        if item.get('sender') == sender:
            for term in item.get('terms'):
                terms.append(term)
    d = dict(zip(terms, [0] * len(terms)))
    for t in terms:
        d[t] += 1
    return d

    
```

```python
blumentritt = get_sender_terms('Ferdinand Blumentritt')
```

```python
len(blumentritt)
```

```python
for t in sorted(blumentritt, key=blumentritt.get, reverse=True):
    print(t, blumentritt[t])
```
